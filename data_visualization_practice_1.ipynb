# 2 Practice basic matplotlib



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%config InlineBackend.figure_format = 'retina'
```

```python
df_mort = pd.read_csv("/Users/victor/Downloads/usa_mortality_rates.csv")

df_mort
```

```python
col_names = list(df_mort.columns[4:].values)
#col_names
```

```python
mort_rate = []
for name in col_names:
    mort_rate.append(df_mort[name][0])
#mort_rate
```

```python
fig, ax = plt.subplots(figsize=(30,10))  # make one subplot (ax) on the figure
ax.bar(col_names, mort_rate, align='center', width=0.6)

ax.set_title("Mortality rate, under-5 (per 1,000 live births)", fontsize=30, loc = 'left')

plt.xlabel("Years", fontsize=20, labelpad=15)
plt.ylabel("Moratity rate", fontsize=20, labelpad=15)
plt.tick_params(labelsize=10)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(30,10))   # make one subplot (ax) on the figure
ax.plot(col_names,mort_rate, linewidth=10)

ax.set_xlabel("Years", fontsize=20)
ax.set_ylabel("Moratity rate", fontsize=20)
ax.set_title("Mortality rate, under-5 (per 1,000 live births)", fontsize=30, loc = 'left')
plt.tick_params(labelsize=10)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(30,10))  # make one subplot (ax) on the figure
ax.scatter(col_names, mort_rate, s=105)

ax.set_xlabel("Years", fontsize=20)
ax.set_ylabel("Moratity rate", fontsize=20)
ax.set_title("Mortality rate, under-5 (per 1,000 live births)", fontsize=30, loc = 'left')
plt.tick_params(labelsize=10)
plt.show()
```

# 3 Practice slopgraph with matplotlib



```python
scores = {'Peers': (85, 91),
 'Culture': (80, 96),
 'Work environment': (76, 75),
 'Leadership': (59, 62),
 'Rewards & recognition': (41, 45),
 'Perf management': (33, 42),
 'Career development': (49, 33)}
```

```python
fig, ax = plt.subplots(figsize=(3,4))
# Let's use 0 as the left-hand side and 1 as the right-hand side
# (below we will set labels to 2014 for 0 and 2015 for 1)
ax.set_xlim(0-.1,1+.1)
ax.set_ylim(10, 100)
# Draw lines and text associated with scores
for c in scores:
    a,b = scores[c]
    color = '#878787'
    if c=='Career development':
        color = '#F46C43'
    ax.plot([0,1], [a,b], 'o-', lw=2, c=color)
    ax.text(0-.05, a, f"{a:.1f}%", color='#878787',
            horizontalalignment='right', verticalalignment='center')
    ax.text(1+.05, b, f"{b:.1f}%", color='#878787',
            horizontalalignment='left', verticalalignment='center')
    ax.text(0-.30, a, c, color='#878787',
            horizontalalignment='right', verticalalignment='center')
# Make the axes look right
plt.text(-1,100 , 'Employee feedback over time', fontsize=12)
#ax.set_title("Employee feedback over time", loc='left', horizontalalignment='right', verticalalignment='top')
plt.text(-0.8,95 , 'Survey category | Percent favorable', color='#878787', fontsize=8)  
ax.spines['bottom'].set_bounds(0, 1)
ax.set_xticks([0,1])
ax.set_xticklabels(['2014','2015'])
ax.set_yticks([])
# Only show the bottom axis
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_linewidth(.5)
plt.show()
```
